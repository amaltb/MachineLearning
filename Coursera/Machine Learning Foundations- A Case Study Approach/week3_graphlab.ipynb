```python
import graphlab
graphlab.canvas.set_target('ipynb')
```

```python
reviews = graphlab.SFrame('data/amazon_baby.csv')
```

```python
reviews.head()
```

```python
reviews = reviews[reviews['rating'] != 3]
```

```python
len(reviews)
```

```python
reviews['rating'] = reviews['rating'] >= 4
```

```python
reviews.head()
```

```python
reviews['word_count'] = graphlab.text_analytics.count_words(reviews['review'])
```

```python
reviews.head()
```

```python
train_data, test_data = reviews.random_split(0.8, seed=0)
```

```python
sentiment_model = graphlab.logistic_classifier.create(train_data, 
                                                      target='rating', 
                                                      features=['word_count'], validation_set=test_data)
```

```python
sentiment_model.evaluate(test_data, metric='roc_curve')
```

```python
sentiment_model.show(view='Evaluation')
```

```python
def word_count(wc, word):
    if word in wc.keys():
        return wc.get(word)
    else:
        return 0
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 
                  'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
for word in selected_words:
    reviews[word] = reviews['word_count'].apply(lambda x: word_count(x, word))
```

```python
reviews.head()
```

```python
for word in selected_words:
    print word + ' : ' + str(reviews[word].sum())
```

```python
train_data_n, tes_data_n = reviews.random_split(.8, seed=0)
```

```python
train_data_n.head()
```

```python
selected_words_model = graphlab.logistic_classifier.create(train_data_n, 
                                                      target='rating', 
                                                      features=selected_words, validation_set=tes_data_n)
```

```python
selected_words_model['coefficients']
```

```python
coeff = selected_words_model['coefficients'].sort('value', ascending=False)
coeff.print_rows(num_rows=12)
```

```python
selected_words_model.evaluate(tes_data_n)
```

```python
selected_words_model.show(view='Evaluation')
```
